```python
import pandas as pd
```

```python
batting_df = pd.read_csv('lahman_data/Batting.csv')

projection_df = batting_df.loc[batting_df.yearID >= 2013]
```

```python
projection_df.info()
```

```python
avgs = []
slgs = []

for index,row in projection_df.iterrows():
    if row.AB == 0:
        avg = 0
        slg = 0
        avgs.append(avg)
        slgs.append(slg)
    else:
        avg = (row.H)/(row.AB)
        avgs.append(avg)

        slg = ((row.H) + (row['2B']) + (2 * row['3B']) + (3 * row.HR))/(row.AB)
        slgs.append(slg)
    
projection_df['AVG'] = avgs
projection_df['SLG'] = slgs

projection_df
```

```python
projection_df['PA'] = (projection_df['AB'] + projection_df['BB'] + projection_df['SH'] + 
                       projection_df['SF'] + projection_df['HBP'])

projection_df.head()
```

```python
df_2016 = projection_df.loc[projection_df.yearID == 2016]
df_2017 = projection_df.loc[projection_df.yearID == 2017]
df_2018 = projection_df.loc[projection_df.yearID == 2018]
df_2019 = projection_df.loc[projection_df.yearID == 2019]
```

```python
projection_df.columns
```

```python
HRs = []
PAs = []

for index, row in df_2016.iterrows():
    hr = (row.HR * 3)
    pa = 200
    HRs.append(hr)
    PAs.append(pa)
    
df_2016['w_HR'] = HRs
df_2016['w_PA'] = PAs

```

```python
HRs = []
PAs = []

for index, row in df_2017.iterrows():
    hr = (row.HR * 4)
    pa = (row.PA * 0.1)
    HRs.append(hr)
    PAs.append(pa)
    
df_2017['w_HR'] = HRs
df_2017['w_PA'] = PAs
```

```python
HRs = []
PAs = []

for index, row in df_2018.iterrows():
    hr = (row.HR * 5)
    pa = (row.PA * 0.5)
    HRs.append(hr)
    PAs.append(pa)
    
df_2018['w_HR'] = HRs
df_2018['w_PA'] = PAs
```

```python
PA_2016 = df_2016.PA.sum()
PA_2017 = df_2017.PA.sum()
PA_2018 = df_2018.PA.sum()

HR_2016 = df_2016.HR.sum()
HR_2017 = df_2017.HR.sum()
HR_2018 = df_2018.HR.sum()

EHR_2016 = HR_2016/PA_2016
EHR_2017 = HR_2017/PA_2017
EHR_2018 = HR_2018/PA_2018
```

```python
#Weighted mean of home runs for someone with X_Players Playing Time

e_hrs = []
e_PAs = []

for player in df_2016.playerID:
    thing = df_2016.loc[df_2016.playerID == player]

    e_hr = ((EHR_2016) * (thing.PA.values[0]) * 3)
    e_pa = (thing.PA.values[0] * 3)
    
    e_PAs.append(e_pa)
    e_hrs.append(e_hr)

df_2016['e_pa'] = e_PAs    
df_2016['e_hr'] = e_hrs

df_2016
```

```python
e_hrs = []
e_PAs = []

for player in df_2017.playerID:
    thing = df_2017.loc[df_2017.playerID == player]

    e_hr = ((EHR_2017) * (thing.PA.values[0]) * 4)
    e_pa = (thing.PA.values[0] * 4)
    
    e_PAs.append(e_pa)
    e_hrs.append(e_hr)

df_2017['e_pa'] = e_PAs    
df_2017['e_hr'] = e_hrs

df_2017
```

```python
e_hrs = []
e_PAs = []

for player in df_2018.playerID:
    thing = df_2018.loc[df_2018.playerID == player]

    e_hr = ((EHR_2018) * (thing.PA.values[0]) * 5)
    e_pa = (thing.PA.values[0] * 5)
    
    e_PAs.append(e_pa)
    e_hrs.append(e_hr)

df_2018['e_pa'] = e_PAs    
df_2018['e_hr'] = e_hrs

df_2018
```

```python
marcel_df = pd.DataFrame()

players = []
expect_hrs = []
expect_pas = []
years = []
weight_hrs = []
weight_pas = []

df_list = [df_2016, df_2017, df_2018]
df_list[0].playerID

for df in df_list:
    for index, row in df.iterrows():
        guy = row.playerID
        players.append(guy)
        
        expect_hr = row['e_hr']
        expect_hrs.append(expect_hr)
        
        expect_pa = row['e_pa']
        expect_pas.append(expect_pa)
        
        year = row['yearID']
        years.append(year)
        
        weight_hr = row['w_HR']
        weight_hrs.append(weight_hr)
        
        weight_pa = row['w_PA']
        weight_pas.append(weight_pa)
    
        
marcel_df['playerID'] = players
marcel_df['yearID'] = years
marcel_df['e_hr'] = expect_hrs
marcel_df['e_pa'] = expect_pas
marcel_df['w_hr'] = weight_hrs
marcel_df['w_pa'] = weight_pas

marcel_df
```

```python
marcel_df_sub = pd.DataFrame()

total_ehrs = []
total_epas = []
guys = []
weights_hrs = []
weight_pas = []

for player in marcel_df.playerID:
#     guy = player
#     guys.append(guy)
    
    mask = marcel_df.loc[marcel_df.playerID == player]
    
    guy = mask.playerID.values[0]
    total_epa = mask['e_pa'].sum()
    total_ehr = mask['e_hr'].sum()
    total_w_hr = mask['w_hr'].sum()
    total_w_pa = mask['w_pa'].sum()
    
    total_epas.append(total_epa)
    total_ehrs.append(total_ehr)
    weights_hrs.append(total_w_hr)
    weight_pas.append(total_w_pa)
    guys.append(guy)
    

marcel_df_sub['playerID'] = guys
marcel_df_sub['e_hr'] = total_ehrs
marcel_df_sub['e_pa'] = total_epas
marcel_df_sub['w_hr'] = weights_hrs
marcel_df_sub['w_pa'] = weight_pas

marcel_df_sub
# t = marcel_df.loc[marcel_df.playerID == 'zuninmi01']
# t.e_hr.sum()
```

```python
hr_perPAs = []

for index, row in marcel_df_sub.iterrows():
    if int(row['e_pa'] == 0):
        per_pa = 0
        hr_perPAs.append(per_pa)
    else:
        regress_hr = (int(row['e_hr']) * 1200) / (int(row['e_pa']))
        per_pa = (regress_hr + int(row['w_hr'])) / (1200 + (int(row['e_pa'])))

        hr_perPAs.append(per_pa)
    
marcel_df_sub['hr_perPA'] = hr_perPAs
marcel_df_sub.head()
```

```python
hrs_2019 = []

for index, row in marcel_df_sub.iterrows():
    ex_hr2019 = ((int(row['w_pa'])) * (float(row['hr_perPA'])))
    hrs_2019.append(ex_hr2019)
    
marcel_df_sub['proj_HR_2019'] = hrs_2019

marcel_df_sub.head()
```

```python
proj_2019 = marcel_df_sub.drop_duplicates()
final_df = proj_2019.sort_values('proj_HR_2019', ascending=False).reset_index(drop=True)
```
